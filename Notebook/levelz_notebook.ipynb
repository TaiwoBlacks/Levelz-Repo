```python
# Import Libraries
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Import raw CSV data to python DataFrame
Data_Dir = Path("/Users/taiwoblackson/Documents/DE Learning/Python/Levelz-Repo/Raw_data/")
df = pd.read_csv(Data_Dir / 'new_retail_data.csv')
print(df.shape)
df.head()
```

```python
print(df.shape)
```

```python
# Check for missing codes
df.info()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["Transaction_ID", "Customer_ID"])
df.shape
```

```python
Category_Columns = ["Gender", "Customer_Segment", "Shipping_Method", "Payment_Method", "Order_Status"]
for col in Category_Columns:
    df[col] = df[col].fillna("unknown")
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
# Adjust data types

# View all the data types using info for .dtypes
df.dtypes
df.info()
```

```python
df["Customer_ID"] = df["Customer_ID"].astype("Int64")
df["Transaction_ID"] = df["Transaction_ID"].astype("Int64")
df.dtypes
```

```python
df["Age"] = df["Age"].astype("Int64")

df["Phone"] = df["Phone"].astype("str").str.replace(".0", "", regex=False)
df["Zipcode"] = df["Zipcode"].astype("str").str.replace(".0", "", regex=False)
df["Year"] = df["Year"].astype("str").str.replace(".0", "", regex=False)
df.dtypes
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors ='coerce')
df.dtypes
df.shape
```

```python
# Dropping Duplicates
df = df.drop_duplicates()

# Ensure no duplicates in 'Transaction_IDs'
df = df.drop_duplicates(subset=['Transaction_ID'])

df.shape
```

```python
# Create new columns

# Average spend per purchase
df["Avg_Spend"] = df["Total_Amount"] / df["Total_Purchases"]

# Customer age groups
df["Age_Group"] = pd.cut(df["Age"],
                         bins = [0, 18, 30, 45, 60, 100],
                         labels = ["teen", "Young Adult", "Adult", "Middle Age", "Senior"])

# Extract weekday from dates
df["Weekday"] = df["Date"].dt.day_name()
```

```python
df.columns
```

```python
df.head()
```

```python
df["High_Value"] = np.where(df["Total_Amount"] > df["Total_Amount"].median(), 1, 0)
df.info()
```

```python
df["Product_Type"]
```

```python
# A place for Numpy: Example of Numpy use here

# Customer-Product ratings matrix (for recommendations) 
Ratings_Matrix = df.pivot_table(index='Customer_ID',
                                columns='Product_Type',
                                values='Ratings',
                                aggfunc='mean').fillna(0)
Ratings_Matrix
```

```python
Ratings_Array = Ratings_Matrix.to_numpy()
print(Ratings_Array)
```

```python
df.columns
```

```python
# Normalize 

# Customers Table
Customers = df[['Customer_ID', 'Name', 'Email', 'Phone', 'Address',
                   'City', 'State', 'Zipcode', 'Country', 'Age', 'Gender',
                   'Income', 'Customer_Segment', 'Age_Group', 'High_Value']
                 ].drop_duplicates(subset=['Customer_ID']).reset_index(drop=True)
Customers
```

```python
# Products table
# Step 1: Select product-related columns and drop duplicates
Products = df[['Product_Category', 'Product_Brand', 'Product_Type']].drop_duplicates().reset_index(drop=True)

# Step 2: Assign unique Product_ID
Products['Product_ID'] = Products.index + 1

# Step 3: Reorder columns to have Product_ID first
Products = Products[['Product_ID'] + [col for col in Products.columns if col != 'Product_ID']]
Products
```

```python
# Orders Table

Orders = df[['Transaction_ID', 'Customer_ID', 'Date', 'Year', 'Month', 'Weekday', 'Time', 'Total_Purchases',
                'Amount', 'Total_Amount', 'Shipping_Method', 'Order_Status', 'products']
                 ].drop_duplicates(subset=['Transaction_ID']).reset_index(drop=True)
Orders
```

```python
Orders['Order_Status']
```

```python
df.columns
```

```python
# Payments table
# Step 1: Select payment-related columns and drop duplicates
Payments = df[['Transaction_ID', 'Amount', 'Payment_Method']].drop_duplicates().reset_index(drop=True)

# Step 2: Assign unique Payment_ID
Payments['Payment_ID'] = Payments.index + 1

# Step 3: Reorder columns to have Payment_ID first
Payments = Payments[['Payment_ID'] + [col for col in Payments.columns if col != 'Payment_ID']]
Payments
```

```python
df.columns
```

```python
# Feedback table
# Step 1: Select payment-related columns and drop duplicates
Feedbacks = df[['Transaction_ID', 'Product_Type', 'Ratings', 'Feedback']].dropna(subset=['Ratings', 'Feedback']).reset_index(drop=True)

# Step 2: Assign unique Payment_ID
Feedbacks['Feedback_ID'] = Feedbacks.index + 1

# Step 3: Reorder columns to have Payment_ID first
Feedbacks = Feedbacks[['Feedback_ID'] + [col for col in Feedbacks.columns if col != 'Feedback_ID']]
Feedbacks
```

```python
# Check for the tables shapes
print('Feedbacks', Feedbacks.shape)
print('Orders', Orders.shape)
print('Products', Products.shape)
print('Customers', Customers.shape)
print('Payments', Payments.shape)
```
